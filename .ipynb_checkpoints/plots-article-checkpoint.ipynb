<h1>Import modules</h1>

```python
# Modules necessary for computation
import math ; pi=math.pi
import numpy as np
import scipy.integrate as integrate
from scipy.interpolate import interp1d
from classy import Class
```

```python
# Optional modules
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

<h1>Compute cosmology and P(k) with CLASS</h1>

```python
# Make a CLASS class
cosmo = Class()
# Set cosmological parameters
ombh2  = 0.022 
omch2  = 0.12
omk    = 0.
hubble = 67.
ns     =  0.96
sig8   =  0.81 
cosmo.set({'omega_b':ombh2,'omega_cdm':omch2,'H0':hubble,'n_s':ns,'sigma8':sig8,'output':'mPk'})
h      = hubble/100. #for  conversions Mpc/h <-> Mpc
# Computation itself
cosmo.compute()
```

<h1>Compute Sij matrix</h1>

<h3>Define arrays of z, r(z), k, P(k)...</h3>

```python
# Redshift bins and array
zstakes    = [0.9,1]
nz_perbin  = 10
nzbins     = len(zstakes)-1
# Redshift and comoving distance arrays
z_arr      = np.zeros((nz_perbin,nzbins))
comov_dist = np.zeros((nz_perbin,nzbins))
for j in range(nzbins):
    z_arr[:,j] = np.linspace(zstakes[j],zstakes[j+1],nz_perbin)
    comov_dist[:,j] = (cosmo.z_of_r(z_arr[:,j]))[0] #In Mpc
# Fourier wavevectors
keq       = 0.02/h                        #Equality matter radiation in 1/Mpc (more or less)
klogwidth = 10                            #Factor of width of the integration range. 10 seems ok ; going higher needs to increase nk_fft to reach convergence (fine cancellation issue noted in Lacasa & Grain)
kmin      = min(keq,1./comov_dist.max())/klogwidth
kmax = max(keq,1./comov_dist.min())*klogwidth
nk_fft    = 2**11                         #seems to be enough. Test precision by increasing, reduce to speed up.
k_4fft    = np.linspace(kmin,kmax,nk_fft) #linear grid on k, as we need to use an FFT
Deltak    = kmax - kmin
Dk        = Deltak/nk_fft
# Power spectrum
Pk_4fft   = np.zeros(nk_fft)
for ik in range(nk_fft):
    Pk_4fft[ik] = cosmo.pk(k_4fft[ik],0.)  #In Mpc^3
# Points where the FFT is computed
dr_fft    = np.linspace(0,nk_fft//2,nk_fft//2+1)*2*pi/Deltak 
```

<h3>Compute necessary FFTs and make interpolation functions</h3>

```python
# n=2
fft2      = np.fft.rfft(Pk_4fft/k_4fft**2)*Dk
dct2      = fft2.real ; dst2 = -fft2.imag
km2Pk_dct = interp1d(dr_fft,dct2,kind='cubic')
# n=3
fft3      = np.fft.rfft(Pk_4fft/k_4fft**3)*Dk
dct3      = fft3.real ; dst3 = -fft3.imag
km3Pk_dst = interp1d(dr_fft,dst3,kind='cubic')
# n=4
fft4      = np.fft.rfft(Pk_4fft/k_4fft**4)*Dk
dct4      = fft4.real ; dst4 = -fft4.imag
km4Pk_dct = interp1d(dr_fft,dct4,kind='cubic')
```

<h3>Compute Sij finally</h3>

```python
Sij = np.zeros((nzbins,nzbins))
for j1 in range(nzbins):
    rmin1   = (comov_dist[:,j1]).min()
    rmax1   = (comov_dist[:,j1]).max()
    zmean1  = ((z_arr[:,j1]).min()+(z_arr[:,j1]).max())/2
    growth1 = cosmo.scale_independent_growth_factor(zmean1)
    pref1   = 3. * growth1 / (rmax1**3-rmin1**3)
    for j2 in range(nzbins):
        rmin2      = (comov_dist[:,j2]).min()
        rmax2      = (comov_dist[:,j2]).max()
        zmean2     = ((z_arr[:,j2]).min()+(z_arr[:,j2]).max())/2
        growth2    = cosmo.scale_independent_growth_factor(zmean2)
        pref2      = 3. * growth2 / (rmax2**3-rmin2**3)
        #p1p2: rmax1 & rmax2
        rsum       = rmax1+rmax2
        rdiff      = abs(rmax1-rmax2)
        rprod      = rmax1*rmax2
        Icp2       = km2Pk_dct(rsum) ; Icm2 = km2Pk_dct(rdiff)
        Isp3       = km3Pk_dst(rsum) ; Ism3 = km3Pk_dst(rdiff)
        Icp4       = km4Pk_dct(rsum) ; Icm4 = km4Pk_dct(rdiff)
        Fp1p2      = -Icp4 + Icm4 - rsum*Isp3 + rdiff*Ism3 + rprod*(Icp2+Icm2)
        #p1m2: rmax1 & rmin2
        rsum       = rmax1+rmin2
        rdiff      = abs(rmax1-rmin2)
        rprod      = rmax1*rmin2
        Icp2       = km2Pk_dct(rsum) ; Icm2 = km2Pk_dct(rdiff)
        Isp3       = km3Pk_dst(rsum) ; Ism3 = km3Pk_dst(rdiff)
        Icp4       = km4Pk_dct(rsum) ; Icm4 = km4Pk_dct(rdiff)
        Fp1m2      = -Icp4 + Icm4 - rsum*Isp3 + rdiff*Ism3 + rprod*(Icp2+Icm2)
        #m1p2: rmin1 & rmax2
        rsum       = rmin1+rmax2
        rdiff      = abs(rmin1-rmax2)
        rprod      = rmin1*rmax2
        Icp2       = km2Pk_dct(rsum) ; Icm2 = km2Pk_dct(rdiff)
        Isp3       = km3Pk_dst(rsum) ; Ism3 = km3Pk_dst(rdiff)
        Icp4       = km4Pk_dct(rsum) ; Icm4 = km4Pk_dct(rdiff)
        Fm1p2      = -Icp4 + Icm4 - rsum*Isp3 + rdiff*Ism3 + rprod*(Icp2+Icm2) 
        #m1m2: rmin1 & rmin2
        rsum       = rmin1+rmin2
        rdiff      = abs(rmin1-rmin2)
        rprod      = rmin1*rmin2
        Icp2       = km2Pk_dct(rsum) ; Icm2 = km2Pk_dct(rdiff)
        Isp3       = km3Pk_dst(rsum) ; Ism3 = km3Pk_dst(rdiff)
        Icp4       = km4Pk_dct(rsum) ; Icm4 = km4Pk_dct(rdiff)
        Fm1m2      = -Icp4 + Icm4 - rsum*Isp3 + rdiff*Ism3 + rprod*(Icp2+Icm2)
        #now group everything
        Fsum       = Fp1p2 - Fp1m2 - Fm1p2 + Fm1m2
        Sij[j1,j2] = pref1 * pref2 * Fsum/(4*pi**2)
```

```python
print Sij
```

<h1>Reproduce results from arXiv:1809.05437</h1>

<h3>Maximum signal to noise</h3>

```python
# Response of the observable, see article
R      = 5
# Maximum signal to noise
SNRmax = 1./(R*np.sqrt(Sij))    
# Knee multipole where SSC dominates.
# Assumes cosmic-variance limited, i.e. negligible shot-noise
ellSSC = np.sqrt(2./(R**2*Sij)) # = sqrt(2)*SNRmax
```

```python
print 'Maximum signal to noise:', SNRmax
print 'SSC multipole:', ellSSC
```

<h3>Signal to noise vs lmax</h3>

<h6>Without any data: cosmic-variance limited case</h6>

```python
# In this case, Gaussian signal to noise is (S/N)_G^2 = sum_ell Cl * (2l+1)/(2*Cl^2) * Cl = sum_ell l+1/2 = lmax^2/2
# and following the article the full signal to noise is (S/N)^2 = (S/N)_G^2 / (1+(S/N)_G^2/(S/N)_max^2)
SNR_max = SNRmax[0]
ell = np.linspace(1,2000)
SNR_G = ell/np.sqrt(2)
SNR_tot = SNR_G/np.sqrt(SNR_G**2/SNR_max**2+1)
plt.plot(ell,SNR_G,lw=3,label='Gaussian')
plt.plot(ell,SNR_tot,lw=3,label='tot')
plt.legend(labelspacing=0.0,fontsize=15,loc='upper left')
plt.xlabel('$\ell_\mathrm{max}$',fontsize=20)
plt.ylabel('S/N',fontsize=20)
plt.show()
```

<h6>With data</h6>

```python
# Load data
dico = np.load("fastSSC.npz")
print dico.keys()
lcenter               = dico['lcenter']               # Center of the bin of multipole
binwidth              = dico['binwidth']              # Width of the bin of multipole
clbin                 = dico['clbin']                 # Binned power spectrum
CovCl_SSC_sigma2_lbin = dico['CovCl_SSC_sigma2_lbin'] # SSC term of binned covariance, full computation
CovCl_SSC_Sij_lbin    = dico['CovCl_SSC_Sij_lbin']       # SSC term of binned covariance, Sij approximation with scale-dependent response R_ell
nbins                 = len(lcenter)
```

```python
# Build Gaussian covariance matrix with usual formula
CovCl_G_lbin = np.zeros((nbins,nbins))
for ilbin in range(0,nbins):
    CovCl_G_lbin[ilbin,ilbin] = 2*clbin[ilbin]**2/((2*lcenter[ilbin]+1)*binwidth[ilbin])
```

```python
# Compute signal to noise ratio depending on lmax
SNR_G_arr   = np.zeros(nbins)
SNR_Sij_arr  = np.zeros(nbins)
SNR_sigma2_arr  = np.zeros(nbins)
for nbinmax in range(0,nbins):
    dvec = np.matrix(clbin[0:nbinmax])
    cGau = np.matrix(CovCl_G_lbin[0:nbinmax,0:nbinmax])
    cSij = np.matrix(CovCl_SSC_Sij_lbin[0:nbinmax,0:nbinmax])
    csigma2 = np.matrix(CovCl_SSC_sigma2_lbin[0:nbinmax,0:nbinmax])
    SNR_G_arr[nbinmax]   = np.sqrt(dvec*np.linalg.inv(cGau)*dvec.transpose())
    SNR_Sij_arr[nbinmax]  = np.sqrt(dvec*np.linalg.inv(cGau+cSij)*dvec.transpose())
    SNR_sigma2_arr[nbinmax]  = np.sqrt(dvec*np.linalg.inv(cGau+csigma2)*dvec.transpose())
```

```python
# Compute Lacasa & Grain formula for the signal to noise ratio 
# with constant effective response R_ell=5 :
# (S/N)^2 = (S/N)_G^2 / (1+Y) with Y=(S/N)_G^2 *R^2 * Sii
R               = 5
SNR_max         = 1./(R*np.sqrt(Sij[0]))
Y               = SNR_G_arr**2/SNR_max**2
SNR_effresp_arr = SNR_G_arr / np.sqrt(1+Y) 
```

```python
#Plot cumulative SNR depending on max bin, for all redshift bins
fig = plt.figure(figsize=(6,4.5))
plt.plot(lcenter,SNR_G_arr,lw=3,label='Gaussian')
plt.plot(lcenter,SNR_effresp_arr,lw=3,label='Sij')
plt.plot(lcenter,SNR_sigma2_arr,lw=3,label='full SSC')
#plt.plot(lcenter,SNR_Sij_arr,lw=3,label='Sij full resp.')
plt.legend(labelspacing=0.0,fontsize=15,loc='upper left')
plt.xlabel('$\ell_\mathrm{max}$',fontsize=20)
plt.ylabel('S/N',fontsize=20)
#plt.savefig("SNR_Clgal_GvseffSijvsfullSSC_varylmax.png",bbox_inches='tight')
plt.show()
```

<h3>Fisher forecasts vs lmax</h3>

```python
# Load data
dclbindparams     = dico['dclbindparams']
latex_param_names = dico['latex_param_names']
param_names       = dico['param_names']
nparams           = len(latex_param_names)
```

```python
#Compute cumulative Fisher for any number of multipole bins
Fisher_G_arr   = np.zeros((nparams,nparams,nbins))
Fisher_Sij_arr  = np.zeros((nparams,nparams,nbins))
Fisher_sigma2_arr  = np.zeros((nparams,nparams,nbins))
for nbinmax in range(0,nbins):
    cGau    = np.matrix(CovCl_G_lbin[0:nbinmax,0:nbinmax])
    cSij    = np.matrix(CovCl_SSC_Sij_lbin[0:nbinmax,0:nbinmax])
    csigma2 = np.matrix(CovCl_SSC_sigma2_lbin[0:nbinmax,0:nbinmax])
    for ip in range(0,nparams):
        dvec1 = np.matrix(dclbindparams[ip,0:nbinmax])
        for jp in range(0,nparams):
            dvec2 = np.matrix(dclbindparams[jp,0:nbinmax])
            Fisher_G_arr[ip,jp,nbinmax]       = dvec1*np.linalg.inv(cGau)*dvec2.transpose()
            Fisher_Sij_arr[ip,jp,nbinmax]     = dvec1*np.linalg.inv(cGau+cSij)*dvec2.transpose()
            Fisher_sigma2_arr[ip,jp,nbinmax]  = dvec1*np.linalg.inv(cGau+csigma2)*dvec2.transpose()
```

```python
# Now compute the costheta defined in Lacasa & Grain
#costheta = <dCldp,V>/sqrt(<dCldp,dCldp><V,V>) with V=Rl*Cl and <X,Y>=X*Cgau^(-1)*Y
costheta_arr   = np.zeros((nparams,nbins))
for nbinmax in range(0,nbins):
    cGau  = np.matrix(CovCl_G_lbin[0:nbinmax,0:nbinmax])
    icGau = np.linalg.inv(cGau)
    for ip in range(0,nparams):
        dvec = np.matrix(dclbindparams[ip,0:nbinmax])
        Vvec = np.matrix(R*clbin[0:nbinmax])
        normd = dvec*icGau*dvec.transpose() ; normV = Vvec*icGau*Vvec.transpose() ; scalprod = dvec*icGau*Vvec.transpose()
        costheta_arr[ip,nbinmax] = scalprod / np.sqrt(normd*normV)
```

```python
# Plot costheta^2, for cosmo parameters only
fig = plt.figure(figsize=(8,6))
for ip in range(5,nparams):
    plt.plot(lcenter,costheta_arr[ip,:]**2,lw=3,label=latex_param_names[ip])
plt.xlim([50,2000]) ; plt.ylim([-0.05,1.05])
plt.legend(labelspacing=0.0,fontsize=15,loc='lower left')
plt.ylabel(r'$\cos^2\theta$',fontsize=20)
plt.xlabel('$\ell_\mathrm{max}$',fontsize=20)
#plt.savefig("cos2theta_vslmax.png",bbox_inches='tight')
plt.show()
```

```python
# Implement Fisher formula from Lacasa & Grain
# with constant effective response R_ell=5 :
# F_p,p = F^G_p,p * (1-cos^2 theta_p * Y/(1+Y))
Fisher_effresp_arr = np.zeros([nparams,nparams,nbins])
for ip in range(0,nparams):
    Fisher_effresp_arr[ip,ip,:] = Fisher_G_arr[ip,ip,:] * (1 - costheta_arr[ip,:]**2 * Y/(1+Y))
```

```python
#Plot cumulative Fisher of some param depending on max bin
fig = plt.figure(figsize=(15,8))
for ip in range(5,nparams):
    ax = plt.subplot(2,3,ip-4)
    plt.plot(lcenter,np.sqrt(Fisher_G_arr[ip,ip,:]),lw=3,label='Gaussian',color='blue')
    plt.plot(lcenter,np.sqrt(Fisher_effresp_arr[ip,ip,:]),lw=3,label='Sij',color='green')
    plt.plot(lcenter,np.sqrt(Fisher_Sij_arr[ip,ip,:]),lw=3,label='Sij w/ '+'$R_\ell$',color='cyan')
    plt.plot(lcenter,np.sqrt(Fisher_sigma2_arr[ip,ip,:]),lw=3,label='full SSC',color='red')
    plt.xlim([50,2000])
    plt.text(0.05, 0.8, latex_param_names[ip], fontsize=20, verticalalignment='bottom', horizontalalignment='left',transform=ax.transAxes)
    if (ip==5):
        plt.legend(labelspacing=0.0,fontsize=15,loc='upper center')
#plt.savefig("Fisher_Clgal_GvseffSijvsfullSSC_varylmax.png",bbox_inches='tight')
plt.show()
```
